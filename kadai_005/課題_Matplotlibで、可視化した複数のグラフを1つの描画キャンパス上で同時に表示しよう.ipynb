```python
import matplotlib.pyplot as plt
import numpy as np
import japanize_matplotlib

fig, ax = plt.subplots(3, 3, figsize=(12, 12),tight_layout=True)
ax = ax.flatten()

# グラフ１
x1 = [0, 1, 2, 3]
y1 = [1, 2, 1, 2]
ax[0].plot (x1, y1)
ax[0].set_title("折れ線グラフ")

#　グラフ２

x2 = np.linspace(0, 10, 30)
y2 = 2 + 2 * np.sin(2 * x2)
ax[1].plot(x2, y2, linewidth=2.0)
ax[1].set(xlim=(0, 10), xticks=np.arange(0, 10),
       ylim=(-1, 5), yticks=np.arange(-1, 6))
ax[1].set_title("曲線グラフ")



# グラフ３

x3 = np.random .normal(15, 5, 2000)
ax[2].set_xlim(-10, 40)
ax[2].hist(x3, bins=30)
ax[2].set_title("ヒストグラム")

# グラフ４

hertz = [300, 200, 200, 500, 600, 2000, 1600, 1800, 1850, 1700, 1500, 1500, 1500, 1400, 5000, 6000, 2000, 3000, 3000]
ax[3].boxplot(hertz)
ax[3].set_title("モーター音の周波数")


# グラフ５

np.random.seed(3)
x4 = 4 + np.random.normal(0, 2, 60)
y4 = 4 + np.random.normal(0, 2, len(x4))
ax[4].scatter(x4, y4)
ax[4].set_title("点プロット図")

# グラフ６

x5 = np.array([30, 40, 30])
label = ["いぬ", "ねこ", "その他"]
ax[5].pie(x5, labels=label)
ax[5].axis("equal")
ax[5].set_title("円グラフ")

# グラフ7

x6 = np.array([1, 2, 3, 4, 5])
y6 = np.array([100, 120, 150, 200, 160])
label = ["昆布", "うめ", "鮭", "カルビ", "すじこ"]
ax[6].bar(x6, y6, tick_label=label)
ax[6].set_title("おにぎりの具ごとの値段")

for i in range(7, len(ax)):
    ax[i].axis('off')

plt.show()
```
